```python
import numpy as np
import pandas as pd
```

```python
# probabilities related to class attributes 
# for laplacian just add 1 to every value of sample and also add to total samples                          
def attributeProb(data, labelAttribute, attributeSize):
    numberofSamples = data.shape[0] +  attributeSize
    classProb = []
    for i in range(data.columns.shape[0]):
        attribute = data.columns.values[i]
        if attribute==labelAttribute:
            continue
        classData = data[attribute].value_counts()
        if( classData.keys().shape[0] != attributeSize ):
            missingKeys = list(set([1,2,3,4,5]) - set(classData.keys()))
            for k in range(len(missingKeys)):
                classData = classData.append(pd.Series(0, index=[missingKeys[k]]))
        classData = classData.sort_index().tolist()
        classData = [(x+1)/numberofSamples for x in classData]
        classProb.append(classData)
    
#     print(classProb)
    return classProb
```

```python
# probabilities of classes are calculated
def calcProbTable(data, labelAttribute, attributeSize):
    numberOfAttributes = data.columns.shape[0]-1                  # one column will be of label
    totalSamples = data.shape[0]
    classes = data[labelAttribute].unique()
    classProb = []
    condAttributeProb =[]
    for i in range(classes.size):
        classProb.append(data[data[labelAttribute]==i].shape[0]/totalSamples)
        condAttributeProb.append(attributeProb(data[data[labelAttribute]==i], labelAttribute, attributeSize))
    return condAttributeProb, classProb   
```

```python
def predictLabel(samplerow, condAttributeProb, classProb, attributeSize):
    samplerow = samplerow.tolist()
    posteriorProb = []
    for i in range(len(classProb)):
        postClass = classProb[i]
        for j in range(attributeSize):
            postClass *= condAttributeProb[i][j][samplerow[j+1]-1]
        posteriorProb.append(postClass)
    print(posteriorProb)
    return posteriorProb.index(max(posteriorProb))    
```

```python
def main():
    data = pd.read_csv("data2_19.csv") 
    labelAttribute = "D"
    attributeSize = 5
    # calculate probability distribution table
    condAttributeProb, classProb = calcProbTable(data, labelAttribute, attributeSize)
    
    # calculate accuracy on test data
    sampleData = pd.read_csv("test2_19.csv")
#     sampleData = data
    correct = 0
    for index, row in sampleData.iterrows():
        if row[labelAttribute]==predictLabel(row, condAttributeProb, classProb, attributeSize):
            correct += 1
    accuracy = (correct/sampleData.shape[0])*100
    print("Accuracy on test data %d" %(accuracy))

if __name__ == '__main__':
    main()
```
